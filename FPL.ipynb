```python
# Allow running nested event loops
nest_asyncio.apply()

async def main():
    # Create a session
    async with aiohttp.ClientSession() as session:
        # Create an instance of FPL with the session
        fpl = FPL(session)

        # Use the await keyword since get_players is an asynchronous method
        potential_players = await fpl.get_players()

        # Now you can convert the data into a DataFrame
        player_dict = [vars(x) for x in potential_players]
        players_df = pd.DataFrame(player_dict)

        return players_df

# Run the asynchronous function in the event loop
players_data = asyncio.run(main())
```

```python
from pulp import LpProblem, LpVariable, lpSum, LpMaximize

# Define data for the existing 15 players
df = raw_data.copy()

# Tworzymy problem optymalizacyjny
prob = LpProblem("FantasyFootball", LpMaximize)

# Tworzymy zmienne decyzyjne dla każdego zawodnika
players_vars = LpVariable.dicts("Player", df.index, cat="Binary")

# Funkcja celu - maksymalizujemy sumę punktów
prob += lpSum(df["total_points"][i] * players_vars[i] for i in df.index)

# Ograniczenie - wybieramy dokładnie 15 zawodników
prob += lpSum(players_vars[i] for i in df.index) == 15

# Ograniczenia dotyczące pozycji zawodników
prob += lpSum(players_vars[i] for i in df.index if df["element_type"][i] == 1) == 2  # Bramkarz
prob += lpSum(players_vars[i] for i in df.index if df["element_type"][i] == 2) == 5  # Obrońcy
prob += lpSum(players_vars[i] for i in df.index if df["element_type"][i] == 3) == 5  # Pomocnicy
prob += lpSum(players_vars[i] for i in df.index if df["element_type"][i] == 4) == 3  # Napastnicy

# Warunek dotyczący budżetu (<1000)
prob += lpSum(df["now_cost"][i] * players_vars[i] for i in df.index) <= 1000

# Rozwiązujemy problem optymalizacyjny
prob.solve()

# Wyświetlamy wyniki
selected_players = [i for i in df.index if players_vars[i].value() == 1]
selected_df = df.loc[selected_players]
```

```python
from pulp import LpProblem, LpMaximize, LpVariable, lpSum, LpInteger

# Załóżmy, że df to twoja tabela z danymi
# Tworzymy model PuLP
prob = LpProblem("FantasyFootball", LpMaximize)

# Zmienna decyzyjna dla każdego gracza (0 lub 1)
x = {i: LpVariable(f"x{i}", 0, 1, LpInteger) for i in df['id']}

# Maksymalizacja punktów
prob += lpSum(df['total_points'] * x[i] for i in df['id'])

# Ograniczenie dotyczące liczby graczy każdego typu
for et, count in zip([1, 2, 3, 4], [2, 5, 5, 3]):
    prob += lpSum(x[i] for i in df[df['element_type'] == et]['id']) >= count, f"Type{et}Constraint"

# Ograniczenie dotyczące kosztu
prob += lpSum(df['now_cost'] * x[i] for i in df['id']) <= 1000, "CostConstraint"

# Ograniczenie dotyczące liczby ogólnych graczy (w tym przypadku 15)
prob += lpSum(x[i] for i in df['id']) == 15, "TotalPlayersConstraint"

# Ograniczenie dotyczące liczby punktujących graczy
prob += lpSum(x[i] for i in df[df['element_type'] == 2]['id']) >= 3, "Type2PlayersConstraint"
prob += lpSum(x[i] for i in df[df['element_type'] == 4]['id']) >= 1, "Type4PlayersConstraint"

# Rozwiązanie problemu
prob.solve()

# Sprawdź wartości zmiennych decyzyjnych (x) w wynikach optymalizacji
selected_players = [i for i in df['id'] if x[i].value() == 1]

# Wyświetl listę identyfikatorów wybranych graczy
print("Selected players:", selected_players)
```
