- players overall dataset

```python
# Allow running nested event loops
nest_asyncio.apply()

async def main():
    # Create a session
    async with aiohttp.ClientSession() as session:
        # Create an instance of FPL with the session
        fpl = FPL(session)

        # Use the await keyword since get_players is an asynchronous method
        potential_players = await fpl.get_players()

        # Now you can convert the data into a DataFrame
        player_dict = [vars(x) for x in potential_players]
        players_df = pd.DataFrame(player_dict)

        return players_df

# Run the asynchronous function in the event loop
players_data = asyncio.run(main())
```

##  Task: Assemble a lineup of 15 players that would currently have the highest points in the season


Rules:  
To join the game select a fantasy football squad of 15 players, consisting of:  

2 Goalkeepers  
5 Defenders  
5 Midfielders  
3 Forwards  
Budget  
The total value of your initial squad must not exceed £100 million.  

Players Per Team  
You can select up to 3 players from a single Premier League team.  

```python
raw_data = players_data[['id','first_name','web_name','now_cost','points_per_game','total_points','team','element_type','points_per_game_rank_type']]
```

```python
# element_type is the player position
raw_data['element_type'].unique()
```

```python
from scipy.optimize import linprog
import numpy as np

df = raw_data.copy()

# Funkcja celu do maksymalizacji
def objective_function(x):
    return sum(df['total_points'] * x)

# Ograniczenie dotyczące kosztu
cost_constraint = {'type': 'ineq', 'fun': lambda x: 1000 - sum(df['now_cost'] * x)}

# Liczba graczy każdego typu
players_count_constraints = [
    {'type': 'eq', 'fun': lambda x, et=et, count=count: sum((df['element_type'] == et) * x) - count} for et, count in zip([1, 2, 3, 4], [2, 5, 5, 3])
]

# Każdy gracz jest wybrany lub nie
bounds = [(0, 1) for _ in range(len(df))]

initial_guess = [0] * len(df)

result = linprog(
    c=[-p for p in df['total_points']],  # Maksymalizacja punktów (uwaga: zmieniłem znak)
    bounds=bounds,  # Każdy gracz jest wybrany lub nie
    method='highs',  # Metoda optymalizacji
    options={'tol': 1e-9},  # Opcjonalne: dokładność rozwiązania
    A_ub=np.array([df['now_cost'].values]),  # Ograniczenie kosztu
    b_ub=[1000],  # Ograniczenie kosztu
    A_eq=np.array([df['element_type'].eq(1).values, df['element_type'].eq(2).values, df['element_type'].eq(3).values, df['element_type'].eq(4).values]),  # Ograniczenia dotyczące typów graczy
    b_eq=[2, 5, 5, 3],  # Liczba graczy każdego typu
)

# Sprawdź wartości zmiennych decyzyjnych (x) w wynikach optymalizacji
selected_players = [df['id'].iloc[i] for i, decision in enumerate(result.x) if decision == 1]

# Wyświetl listę identyfikatorów wybranych graczy
print("Selected players:", selected_players)
```

```python
raw_data[raw_data.id.isin(selected_players)].sort_values('element_type')
```

```python
from pulp import LpProblem, LpVariable, lpSum, LpMaximize

# Define data for the existing 15 players
df = raw_data.copy()

# Tworzymy problem optymalizacyjny
prob = LpProblem("FantasyFootball", LpMaximize)

# Tworzymy zmienne decyzyjne dla każdego zawodnika
players_vars = LpVariable.dicts("Player", df.index, cat="Binary")

# Funkcja celu - maksymalizujemy sumę punktów
prob += lpSum(df["total_points"][i] * players_vars[i] for i in df.index)

# Ograniczenie - wybieramy dokładnie 15 zawodników
prob += lpSum(players_vars[i] for i in df.index) == 15

# Ograniczenia dotyczące pozycji zawodników
prob += lpSum(players_vars[i] for i in df.index if df["element_type"][i] == 1) == 2  # Bramkarz
prob += lpSum(players_vars[i] for i in df.index if df["element_type"][i] == 2) == 5  # Obrońcy
prob += lpSum(players_vars[i] for i in df.index if df["element_type"][i] == 3) == 5  # Pomocnicy
prob += lpSum(players_vars[i] for i in df.index if df["element_type"][i] == 4) == 3  # Napastnicy

# Warunek dotyczący budżetu (<1000)
prob += lpSum(df["now_cost"][i] * players_vars[i] for i in df.index) <= 1000

# Rozwiązujemy problem optymalizacyjny
prob.solve()

# Wyświetlamy wyniki
selected_players = [i for i in df.index if players_vars[i].value() == 1]
selected_df = df.loc[selected_players]
```
